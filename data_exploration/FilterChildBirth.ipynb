## Split the cleaned dataset into two data frames according to the value of column "Birth Weight":
* SPARCS-2017-birth.csv: records that has "Birth Weight". Same strcuture as the cleaned csv (224082 rows \* 25 cols). According to "CCS Diagnosis Description", not only contains Liveborn, but also contains liveborn-related disease. 
* SPARCS-2017-Notbirth.csv: records that has Null "Birth Weight". Similar strcuture to the cleaned csv but dropped the column "Birth Weight" (2080234 rows \* 24 col).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 

filename = '/Users/mac/Desktop/BigData_ProjectData/SPARCS-2017.csv'
```

```python
data = pd.read_csv(filename)
print(np.shape(data))
data.head()
```

```python
data2 = data.dropna(subset = ['Hospital Service Area','Zip Code - 3 digits',
                             'APR Risk of Mortality'])
```

```python
droplist = ["Discharge Year","Abortion Edit Indicator",
            "Operating Certificate Number","Permanent Facility Id",
            "CCS Diagnosis Code","CCS Procedure Code",
            "APR MDC Code","APR Severity of Illness Code",
           "APR DRG Code"]
data3 = data2.drop(droplist, axis=1)
print(np.shape(data3))
data3.head()
```

```python
data3[data3["Birth Weight"].notnull()&data3['CCS Diagnosis Description']!='Liveborn']#.loc[:, 'Patient Disposition':'Birth Weight']
```

```python
birth = data3[data3["Birth Weight"].notnull()]
not_birth = data3[data3["Birth Weight"].isna()]
print(np.shape(birth),np.shape(not_birth))
```

```python
set(birth["CCS Diagnosis Description"])
```

```python
"Livebirth" in set(not_birth["CCS Diagnosis Description"])
```

```python
birthname = '/Users/mac/Desktop/BigData_ProjectData/SPARCS-2017-birth.csv'
birth.to_csv(birthname, sep='\t', index=False)
```

```python
not_birth = not_birth.drop(["Birth Weight"], axis=1)
print(np.shape(not_birth))
not_birth.count()
```

```python
notname = '/Users/mac/Desktop/BigData_ProjectData/SPARCS-2017-Nobirth.csv'
not_birth.to_csv(notname, sep='\t', index=False)
```
